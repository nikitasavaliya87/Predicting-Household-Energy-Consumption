```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/nikit/OneDrive/Desktop/data/household_power_consumption.txt', sep=';',
                  parse_dates={'Date_Time' : ['Date', 'Time']}, infer_datetime_format=True, 
                 low_memory=False, na_values=['nan','?'],index_col='Date_Time')
```

```python
df.head()
```

```python
df.ffill(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['power_consumption'] = (df['Global_active_power']*1000/60) - (df['Sub_metering_1'] + df['Sub_metering_2'] + df['Sub_metering_3'])
df.head()
```

```python
df['Date'] = df.index.date
df['time'] = df.index.time
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df_date = df.groupby(['Date']).sum()
```

```python
df_power_consumption = df_date[['power_consumption']]
```

```python
df_power_consumption.head()
```

```python
df_power_consumption.groupby(df_power_consumption.index.year).describe()
```

```python
df_power_consumption.groupby(df_power_consumption.index.year).sum().plot(kind="bar",xlabel='Year',ylabel='power consumption',title="Yearly - Total Power Consumption", figsize=(18,6))
```
